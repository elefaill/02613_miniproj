```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the CSV file containing floorplan temperature data
# Replace 'floorplans.csv' with the actual file path
data = pd.read_csv('test10.csv')
```

```python
# a) Distribution of the mean temperatures
# Strip leading/trailing spaces from column names
data.columns = data.columns.str.strip()

plt.hist(data['mean_temp'], bins=5, color='blue', alpha=0.7)
plt.title('Distribution of Mean Temperatures')
plt.xlabel('Mean Temperature (ºC)')
plt.ylabel('Frequency')
plt.show()
```

```python

# b) Average mean temperature of the buildings
average_mean_temperature = data['mean_temp'].mean()
print(f"Average mean temperature of the buildings: {average_mean_temperature:.2f}ºC")

# c) Average temperature standard deviation
temperature_std = data['std_temp'].mean()
print(f"Average temperature standard deviation: {temperature_std:.2f}ºC")
```

```python
# d) Buildings with at least 50% of their area above 18ºC
buildings_above_18 = data[data['pct_above_18'] >= 50]
num_buildings_above_18 = len(buildings_above_18)
print(f"Number of buildings with at least 50% of their area above 18ºC: {num_buildings_above_18}")

# e) Buildings with at least 50% of their area below 15ºC
buildings_below_15 = data[data['pct_below_15'] >= 50]
num_buildings_below_15 = len(buildings_below_15)
print(f"Number of buildings with at least 50% of their area below 15ºC: {num_buildings_below_15}")
```
